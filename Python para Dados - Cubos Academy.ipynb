# Operações com texto

**Operador (+)**


Concatena uma string a outra string, retornando a junção das strings.

```python
qtd_frutas = "2"
fruta1 = "morango"
fruta2 = "banana"

print("Eu como " + qtd_frutas + " frutas pela manhã: " + fruta1 + " e " + fruta2)
```

**f-strings**


Operador que permite colocar uma variável dentro de um texto

```python
qtd_frutas = 2
fruta1 = "maça"
fruta2 = "banana"

print(f"Eu como {qtd_frutas} frutas pela manhã: {fruta1} e {fruta2}!")
```

**len()**

Retorna a quantidade de caracteres em uma string.

```python
nome_completo = "Yasmim Vitoria Silva Vieira"

#Calculando o comprimento
print(len(nome_completo))
```

**upper()    |    lower()**


Upper - altera os caracteres para maiúsculas

Lower - altera os caracteres para minúsculas

```python
nome_completo = "Yasmim Vitoria Silva Vieira"

print(nome_completo.upper())
print(nome_completo.lower())

#Alterar variável
nome_completo = nome_completo.upper()
print(nome_completo)
```

**replace()**

Substitui string especificadas por novas strings.

**Deixar todas as letras maiusculas ou minusculas, antes da alteração!**

```python
txt = "Eu gosto de comer no café da manhã pão e fruta."

print(txt.replace("fruta", "banana"))

#Alterar variável
txt = print(txt.replace("pão", "pão com ovo"))
```

```python
txt = "Eu gosto de comer no café da manhã pão com ovo e banana."

print("pão" in txt)
```
